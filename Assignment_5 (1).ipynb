Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), find

i. Sum of all elements

ii. Sum of all elements row-wise

iii. Sum of all elements column-wise

```python
import numpy as np
mat = np.matrix('[4,1,9;12,3,1;4,5,6]')
print("Matrix =")
print(mat)
print()

s = mat.sum()
row_sum = mat.sum(axis = 1)
col_sum = mat.sum(axis = 0)

print("Sum of all elements =",s)
print("Sum of all elements row-wise =")
print(row_sum)
print("Sum of all elements column-wise =",col_sum)
```

Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting.

a) Generate your unique sales dataset:

Take the sum of the ASCII values of the initials of your first and last

name. Call this value X. (If your initials are A B → ASCII sum = 65

66 = 131 → sales = [131, 181, 231, 281, 331].)

Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200].

```python
first_name=input("enter your first name\n")
last_name=input("enter your last name\n")

X=ord(first_name[0])+ord(last_name[0])
print("Sum of ascii's is ",X)
```

```python
def createarray(X):
    array=np.array([X,X+50,X+100,X+150,X+200])
    return array
array=createarray(X)
```

b) Compute your personalized tax rate as ((X % 5) + 5) / 100.

Use broadcasting to apply this tax rate to each sales value.

```python
def taxRate(X):
    return ((X%5)+5)/100
tax=taxRate(X)

taxed_sales = array * (1+tax)
print(taxed_sales)
```

c) Adjust sales based on discount:

If sales < X+100, apply a 5% discount.

If sales >= X+100, apply a 10% discount.

```python
def discounted_sales(array,X):
    discounted_array=np.where(array<X+100,array*0.95,array*0.90)
    return discounted_array

print(discounted_sales(array,X))
```

d) Expand sales data for multiple weeks:

Create a 3×5 matrix representing three weeks of sales by stacking

sales three times using broadcasting.

Increase sales by 2% per week using element-wise broadcasting.

```python
weekly_sales = np.vstack([array] * 3)  

print(weekly_sales)

weeks = np.arange(3).reshape(3, 1) 
weekly_sales = weekly_sales * (1.02 ** weeks)  

print("Weekly Sales Data:\n", weekly_sales)
```
